```python
import numpy as np
```

```python
def calculate_information_metrics(p_true, p_model):
    """
    Calculates Entropy, Cross-Entropy, and KL Divergence (Bit-Waste).
    All calculations are in bits (log base 2).
    """
    # Convert to numpy arrays for vector math
    y = np.array(p_true)
    theta_hat = np.array(p_model)

    # Invariant check: Ensure distributions sum to 1
    assert np.isclose(y.sum(), 1.0), "True distribution must sum to 1"
    assert np.isclose(theta_hat.sum(), 1.0), "Model distribution must sum to 1"

    # Avoid log(0) by using a tiny epsilon where probability is 0
    epsilon = 1e-15
    y_safe = np.clip(y, epsilon, 1.0)
    theta_safe = np.clip(theta_hat, epsilon, 1.0)

    # 1. Entropy H(y): The theoretical minimum bits
    entropy = -np.sum(y * np.log2(y_safe))

    # 2. Cross-Entropy H(y, theta_hat): Average bits using model's scheme
    cross_entropy = -np.sum(y * np.log2(theta_safe))

    # 3. KL Divergence: The "Waste" (Cross-Entropy - Entropy)
    kl_divergence = cross_entropy - entropy

    return entropy, cross_entropy, kl_divergence
```

```python
# --- Experiment Scenarios ---

# Scenario 1: Perfect Model
# Reality: [Sunny: 80%, Cloudy: 10%, Rainy: 10%]
true_weather = [0.8, 0.1, 0.1]
perfect_model = [0.8, 0.1, 0.1]

# Scenario 2: "Wrong" Model (Thinks Sunny is rare)
wrong_model = [0.2, 0.4, 0.4]

# Run calculations
h_ideal, h_cross_ideal, waste_ideal = calculate_information_metrics(
    true_weather, perfect_model
)
h_true, h_cross_wrong, waste_wrong = calculate_information_metrics(
    true_weather, wrong_model
)

print(f"--- Perfect Model ---")
print(f"Min bits needed (Entropy): {h_ideal:.3f}")
print(f"Bits used by model:        {h_cross_ideal:.3f}")
print(f"Waste (KL Divergence):     {waste_ideal:.3f} bits\n")

print(f"--- Wrong Model ---")
print(f"Min bits needed (Entropy): {h_true:.3f}")
print(f"Bits used by model:        {h_cross_wrong:.3f}")
print(f"Waste (KL Divergence):     {waste_wrong:.3f} bits")
```
