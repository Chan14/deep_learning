```python
import numpy as np
```

```python
W = np.array([[3, 1], [5, 7], [2, 5]])
W_T = W.T
W_T__W = np.matmul(W_T, W)
print(f"W = \n{W}")
print(f"W_T = \n{W_T}")
print(f"W_T__W = \n{W_T__W}")
```

```python
x = np.array([1, 2]).reshape(-1, 1)
print(x)
```

```python
z = np.matmul(W_T__W, x)
print(z)
```

```python
eigenvalues, eigenvectors = np.linalg.eigh(W_T__W)
# Print the results
print("Eigenvalues:", eigenvalues)
print("\nEigenvectors:\n", eigenvectors)

for col in eigenvectors.T:
    print(np.linalg.norm(col))
# 1.0
# 1.0
```

```python
x = np.array([2, 1]).reshape(-1, 1)
y = np.matmul(W_T__W, x)

P = eigenvectors  # columns are eigenvectors
x_in_eigenspace = P.T @ x

print("y = W_T__W @ x:\n", y)
print("\nx in eigenspace:\n", x_in_eigenspace)
```

```python
Lambda = np.diag(eigenvalues)
y_via_eigenspace = P @ Lambda @ x_in_eigenspace
print("\ny reconstructed via eigenspace:\n", y_via_eigenspace)
```

```python
U, S, V_T = np.linalg.svd(W)
print("U:\n", U)
print("\nS:", S)
print("\nV_T:\n", V_T)
print(f"Eigenvectors :\n{eigenvectors}")
W__W_T = np.dot(W, W_T)
print(f"W__W_T = \n{W__W_T}")
eigen_val2, eigen_vec2 = np.linalg.eigh(W__W_T)
print(f"eigen_val2 = {eigen_val2}")
print(f"eigen_vect2 = {eigen_vec2}")
```

```python
x = np.array([2, 1]).reshape(-1, 1)
z = np.dot(W, x)
print(f"z = \n{z}")
Sigma = np.zeros((W.shape[0], W.shape[1]))
np.fill_diagonal(Sigma, S)
z_via_SVD = U @ Sigma @ V_T @ x
print(f"z_via_SVD = \n{z_via_SVD}")
```
