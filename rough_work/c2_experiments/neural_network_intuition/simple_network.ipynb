```python
import numpy as np
```

```python
X_train = np.array([[1, 3, 1], [3, 1, 2]])
y_train = np.array([[9, 11, 7]])
print(f"X_train=\n{X_train}")
print(f"y_train=\n{y_train}")
```

```python
# 4, 2, 1=> 4, 2     2, 4.     1, 2
W1 = np.array([[1, 2], [2, 1], [-1, 2], [-2, 1]])
b1 = np.array([[0], [0], [0], [0]])
W2 = np.array([[1, -1, 2, 3], [-2, 1, 1, 2]])
b2 = np.array([[0], [0]])
W3 = np.array([[1, -1]])
b3 = np.array([[0]])
```

```python
print(f"W1=\n{W1}")
print(f"b1=\n{b1}")
print(f"W2=\n{W2}")
print(f"b2=\n{b2}")
print(f"W3=\n{W3}")
print(f"b3=\n{b3}")
```

```python
Z1 = W1 @ X_train + b1
A1 = np.maximum(0, Z1)
Z2 = W2 @ A1 + b2
A2 = np.maximum(0, Z2)
Z3 = W3 @ A2 + b3
A3 = Z3
print(f"Z1=\n{Z1}")
print(f"A1=\n{A1}")
print(f"Z2=\n{Z2}")
print(f"A2=\n{A2}")
print(f"Z3=\n{Z3}")
print(f"A3=\n{A3}")
```

```python
m = 3
loss = np.sum((Z3 - y_train) ** 2) / (2 * m)
print(f"Loss = {loss}")
```

```python
dZ3 = (Z3 - y_train) / m
dW3 = dZ3 @ A2.T
db3 = np.sum(dZ3, axis=1, keepdims=True)
dA2 = W3.T @ dZ3
dZ2 = dA2 * (Z2 > 0)
dW2 = dZ2 @ A1.T
db2 = np.sum(dZ2, axis=1, keepdims=True)
dA1 = W2.T @ dZ2
dZ1 = dA1 * (Z1 > 0)
dW1 = dZ1 @ X_train.T
db1 = np.sum(dZ1, axis=1, keepdims=True)
```

```python
print(f"dZ3=\n{dZ3}")
print(f"dW3=\n{dW3}")
print(f"db3=\n{db3}")
print(f"dA2=\n{dA2}")
print(f"dZ2=\n{dZ2}")
print(f"dW2=\n{dW2}")
print(f"db2=\n{db2}")
print(f"dA1=\n{dA1}")
print(f"dZ1=\n{dZ1}")
print(f"dW1=\n{dW1}")
print(f"db1=\n{db1}")
```
