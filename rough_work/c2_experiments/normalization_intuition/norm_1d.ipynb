```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_input(ax, X):
    if X.ndim == 1:
        ax.scatter(X, np.zeros_like(X), alpha=0.7, s=60)
    else:
        ax.scatter(X[:, 0], X[:, 1], alpha=0.7, s=60)
    ax.set_xlabel("$X_1$")
    # ax.set_yticks([])  # y-axis is meaningless, hide it
    if X.ndim == 2:
        ax.set_ylabel("$X_2$")
```

```python
X = np.array(range(1, 21))
print(f"X = {X}")
fig, ax = plt.subplots(1, 3, figsize=(16, 6))
plot_input(ax[0], X)
ax[0].set_title("Original inputs")
X_mean = np.mean(X)
X_shifted = X - X_mean
ax[1].set_title("Input centered around 0")
plot_input(ax[1], X_shifted)
X_std = np.std(X_shifted)
X_normal = X_shifted / X_std
plot_input(ax[2], X_normal)
ax[2].set_title("Normalized input")
fig.suptitle("Input Distribution")
```

```python
X1 = np.array(range(1, 21))
X2 = np.array(range(10, 70, 3))
X = np.c_[X1, X2]
print(f"X = {X}")
fig, ax = plt.subplots(1, 3, figsize=(16, 6))
plot_input(ax[0], X)
ax[0].set_title("Original inputs")
X_mean = np.mean(X, axis=0)
print(X_mean)
X_shifted = X - X_mean
plot_input(ax[1], X_shifted)
ax[1].set_title("Input centered around 0")
X_std = np.std(X_shifted, axis=0)
X_normal = X_shifted / X_std
plot_input(ax[2], X_normal)
ax[2].set_title("Normalized input")
fig.suptitle("Input Distribution")
```
