```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider, Dropdown


# Activations
def relu(z):
    return np.maximum(0, z)


def relu_deriv(z):
    return int(z > 0)


def sigmoid(z):
    return 1 / (1 + np.exp(-np.clip(z, -500, 500)))


def sigmoid_deriv(z):
    s = sigmoid(z)
    return s * (1 - s)


# Forward + gradient computation
def forward_and_gradients(activation="relu", bias=0.0):
    # Inputs and target
    x1, x2 = 1.0, 1.0
    target = 0.0

    # Weights chosen for ReLU flip
    w11, w21 = 1.0, 0.5
    w12, w22 = 1.0, 1.0
    v1, v2 = 4.0, -6.0  # opposite signs

    # Pre-activations with shared bias
    z1 = w11 * x1 + w21 * x2 + bias
    z2 = w12 * x1 + w22 * x2 + bias

    # Activations and derivatives
    if activation == "sigmoid":
        h1 = sigmoid(z1)
        h2 = sigmoid(z2)
        deriv1 = sigmoid_deriv(z1)
        deriv2 = sigmoid_deriv(z2)
    else:  # relu
        h1 = relu(z1)
        h2 = relu(z2)
        deriv1 = relu_deriv(z1)
        deriv2 = relu_deriv(z2)

    # Output and delta
    y_hat = v1 * h1 + v2 * h2
    delta_out = y_hat - target

    # Gradient w.r.t shared bias
    grad_b_path1 = delta_out * v1 * deriv1 * 1
    grad_b_path2 = delta_out * v2 * deriv2 * 1
    total_grad_b = grad_b_path1 + grad_b_path2

    # Display info
    print(f"Activation: {activation.upper()}")
    print(f"Bias: {bias:.2f}")
    print(f"z1: {z1:.3f}, z2: {z2:.3f}")
    print(f"deriv1: {deriv1}, deriv2: {deriv2}")
    print(f"Total grad b: {total_grad_b:.3f}")
    print("-" * 40)

    # Bar plot
    labels = ["Path 1", "Path 2", "Total"]
    values = [grad_b_path1, grad_b_path2, total_grad_b]
    colors = ["red" if v < 0 else "green" for v in values]
    plt.figure(figsize=(8, 4))
    plt.bar(labels, values, color=colors)
    plt.title(f"Gradients on shared bias ({activation.upper()})")
    plt.ylabel("Gradient Value")
    plt.ylim(min(values) - 1, max(values) + 1)
    plt.show()


# Interactive widget
@interact(
    activation=Dropdown(
        options=["relu", "sigmoid"], value="relu", description="Activation:"
    ),
    bias=FloatSlider(
        min=-3.0, max=3.0, step=0.05, value=0.0, description="Shared Bias:"
    ),
)
def run_example(activation, bias):
    forward_and_gradients(activation, bias)
```
