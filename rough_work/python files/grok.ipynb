```python
# Simple Neural Network Example: ReLU vs Sigmoid Gradients
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider, Dropdown


def sigmoid(z):
    return 1 / (1 + np.exp(-np.clip(z, -500, 500)))


def sigmoid_deriv(z):
    s = sigmoid(z)
    return s * (1 - s)


def relu(z):
    return np.maximum(0, z)


def relu_deriv(z):
    return int(z > 0)


def forward_and_gradients(activation="relu", perturb_w11=0.0):
    # Setup
    x1, x2 = 1.0, 1.0
    target = 0.0

    w11 = 0.2 + perturb_w11  # The early weight we're watching
    w21 = 1.4
    w12 = -1.5
    w22 = 1.2

    v1, v2 = 7.0, -8.0

    # Pre-activations
    z1 = w11 * x1 + w21 * x2
    z2 = w12 * x1 + w22 * x2

    if activation == "sigmoid":
        h1 = sigmoid(z1)
        h2 = sigmoid(z2)
        deriv1 = sigmoid_deriv(z1)
        deriv2 = sigmoid_deriv(z2)
    else:  # relu
        h1 = relu(z1)
        h2 = relu(z2)
        deriv1 = relu_deriv(z1)
        deriv2 = relu_deriv(z2)

    y_hat = v1 * h1 + v2 * h2
    delta_out = y_hat - target

    # Gradients to hidden
    delta_h1 = delta_out * v1 * deriv1
    delta_h2 = delta_out * v2 * deriv2

    # Gradients to w11 from each path
    grad_w11_path1 = delta_h1 * x1
    grad_w11_path2 = delta_h2 * x1
    total_grad_w11 = grad_w11_path1 + grad_w11_path2

    # Display
    print(f"Activation: {activation.upper()}")
    print(f"Perturb w11 by: {perturb_w11}")
    print(f"z1: {z1:.3f}, h1: {h1:.3f}, deriv1: {deriv1:.3f}")
    print(f"z2: {z2:.3f}, h2: {h2:.3f}, deriv2: {deriv2:.3f}")
    print(f"y_hat: {y_hat:.3f}, delta_out: {delta_out:.3f}")
    print(f"delta_h1: {delta_h1:.3f}, delta_h2: {delta_h2:.3f}")
    print(f"Grad w11 from path1: {grad_w11_path1:.3f}")
    print(f"Grad w11 from path2: {grad_w11_path2:.3f}")
    print(f"Total grad w11: {total_grad_w11:.3f}")
    print("-" * 40)

    # Simple bar plot
    labels = ["Path 1", "Path 2", "Total"]
    values = [grad_w11_path1, grad_w11_path2, total_grad_w11]
    colors = ["red" if v < 0 else "green" for v in values]
    plt.figure(figsize=(8, 4))
    plt.bar(labels, values, color=colors)
    plt.title(f"Gradients on w11 ({activation.upper()})")
    plt.ylabel("Gradient Value")
    plt.ylim(min(values) - 1, max(values) + 1)
    plt.show()


# Interactive widget
@interact(
    activation=Dropdown(
        options=["relu", "sigmoid"], value="relu", description="Activation:"
    ),
    perturb_w11=FloatSlider(
        min=-2.0, max=2.0, step=0.01, value=0.0, description="Perturb w11:"
    ),
)
def run_example(activation, perturb_w11):
    forward_and_gradients(activation, perturb_w11)
```
